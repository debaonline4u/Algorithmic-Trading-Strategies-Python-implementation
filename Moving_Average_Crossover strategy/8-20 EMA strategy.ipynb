```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.offline
import talib as ta
```

```python
import cufflinks as cf
cf.go_offline()
cf.set_config_file(offline=False, world_readable=True)
```

```python
# Load minutes data
path1 = '/home/debashis/Data/Finance_Data/kaggle data/2017_18_19_20/'
df_bn = pd.read_csv(path1+'NIFTY_BANK__EQ__INDICES__NSE__MINUTE.csv', low_memory=False)
df_bn.fillna(method='ffill', inplace=True)
df_bn.head()
```

```python
cols = ['close', 'ema8', 'ema20', 'ema50', 'ema200', 'signal']
df_temp = pd.DataFrame(columns=cols)
# df_temp['close'] = df_bn[['close']].iloc[:1000, :]
df_temp['close'] = df_bn.close.values[:1000]
df_temp.head()
```

```python
# df_reliance[['close']].iplot(kind='line')
```

```python
ema_8 = ta.EMA(df_bn.close.values[:1000], 8)
ema_20 = ta.EMA(df_bn.close.values[:1000], 20)
ema_50 = ta.EMA(df_bn.close.values[:1000], 50)
ema_200 = ta.EMA(df_bn.close.values[:1000], 200)
df_temp['ema8'] = ema_8
df_temp['ema20'] = ema_20
df_temp['ema50'] = ema_50
df_temp['ema200'] = ema_200
```

```python
# df_temp.dropna(inplace=True)
df_temp.head(20)
```

```python
# df_temp.fillna(0, inplace=True)
# df_temp.head(20)
```

```python
signals = []
for index, row in df_temp.iterrows():
    if row['ema8'] > row['ema20']:
        signals.append(1)
    else:
        signals.append(0)
        
print('Signals: ', signals)
```

```python
df_temp.signal = signals
```

```python
# df_temp[['close', 'ema8', 'ema20', 'signal']].iloc[:500].iplot(kind='line')
df_temp[['ema50', 'signal']].iplot(secondary_y='signal', title='Signals and Close price', xTitle='Time in minutes', 
                                   yTitle='Bank nifty price')
# df_temp[['ema8']].iplot(kind='line')
# df_temp[['ema20']].iplot(kind='line')
```
